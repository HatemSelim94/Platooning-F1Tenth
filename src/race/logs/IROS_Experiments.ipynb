```python
import pandas as pd
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
sns.set_theme(style="white")
import os
import plotly.express as px
from plotly.subplots import make_subplots
from plotly import graph_objects as go
import warnings
warnings.filterwarnings("ignore")
```

```python
def rename_controlers(row):
    row = row.strip()
    dicta = {"e2e":"End-to-End","pure_pursuit":"Pure Pursuit","disparity_extender":"Disparity Extender"}
    return dicta[row]
```

```python
two_car_experiments = pd.read_csv('safety_experiments.csv',header=None).rename(columns={0:"Experiment Number",1:"Controller",2:"Velocity",3:"Safety Percentage"})
two_car_experiments
```

```python
two_car_experiments.groupby(["Controller","Velocity"]).head(15).drop(columns="Experiment Number")
two_car_experiments["Controller"] = two_car_experiments["Controller"].apply(rename_controlers)
two_car_experiments
```

```python
results=two_car_experiments.groupby(["Controller","Velocity"]).head(15).drop(columns="Experiment Number").groupby(["Controller","Velocity"]).agg(['mean', 'median','std'])*100
round(results,3)
```
